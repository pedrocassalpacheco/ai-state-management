# Aurora RDS MySQL Connection Test

This notebook tests the connection to the Aurora RDS MySQL instance.

```python
# Import required libraries
import pymysql
import os
from dotenv import load_dotenv
from datetime import datetime
```

## Load Environment Variables

```python
# Load environment variables from .env file
load_dotenv()

DB_HOST = os.getenv('DB_HOST')
DB_PORT = int(os.getenv('DB_PORT', 3306))
DB_USER = os.getenv('DB_USER')
DB_PASSWORD = os.getenv('DB_PASSWORD')
DB_NAME = os.getenv('DB_NAME')

print(f"Database Host: {DB_HOST}")
print(f"Database Port: {DB_PORT}")
print(f"Database User: {DB_USER}")
print(f"Database Name: {DB_NAME}")
print("✓ Environment variables loaded")
```

## Test Aurora RDS MySQL Connection

```python
# Connect to Aurora RDS MySQL
try:
    connection = pymysql.connect(
        host=DB_HOST,
        port=DB_PORT,
        user=DB_USER,
        password=DB_PASSWORD,
        #database=DB_NAME,
        cursorclass=pymysql.cursors.DictCursor,
        connect_timeout=10
    )
    print("✓ Successfully connected to Aurora RDS MySQL")
except Exception as e:
    print(f"✗ Connection failed: {e}")
    raise
```

## Test Basic Queries

```python
# Get MySQL version
with connection.cursor() as cursor:
    cursor.execute("SELECT VERSION() as version")
    result = cursor.fetchone()
    print(f"MySQL Version: {result['version']}")
```

```python
# Show current database
with connection.cursor() as cursor:
    cursor.execute("SELECT DATABASE() as current_db")
    result = cursor.fetchone()
    print(f"Current Database: {result['current_db']}")
```

```python
# Show all databases
with connection.cursor() as cursor:
    cursor.execute("SHOW DATABASES")
    databases = cursor.fetchall()
    print("Available Databases:")
    for db in databases:
        print(f"  - {db['Database']}")
```

```python
# Show tables in current database
with connection.cursor() as cursor:
    cursor.execute("SHOW TABLES")
    tables = cursor.fetchall()
    if tables:
        print("Tables in current database:")
        for table in tables:
            table_name = list(table.values())[0]
            print(f"  - {table_name}")
    else:
        print("No tables found in current database")
```

```python
# Get connection information
with connection.cursor() as cursor:
    cursor.execute("SELECT USER() as user, CONNECTION_ID() as connection_id")
    result = cursor.fetchone()
    print(f"Connected as: {result['user']}")
    print(f"Connection ID: {result['connection_id']}")
```

## Test Simple Query

```python
# Test a simple SELECT query
with connection.cursor() as cursor:
    cursor.execute("SELECT 'Hello from Aurora!' as message, NOW() as timestamp")
    result = cursor.fetchone()
    print(f"Message: {result['message']}")
    print(f"Timestamp: {result['timestamp']}")
```

## Close Connection

```python
# Close the connection
connection.close()
print("✓ Connection closed")
```
